    <h2>مقدمه</h2>
    <p>در این تمرین هدف اصلی شما یادگیری و پیاده‌سازی تکنیک‌های تبدیل هندسی روی تصاویر دیجیتال است. این فرآیند شامل جابجایی و تغییر شکل پیکسل‌های تصویر برای انتقال آن به یک قالب جدید است. از این نوع تبدیل‌ها می‌توان در موارد متعددی مانند اصلاح زاویه دید، هم‌تراز کردن تصاویر برای تحلیل بیشتر، و یا افزودن جلوه‌های بصری خاص استفاده کرد.</p>
    <p>در ادامه قصد داریم از نوع خاصی از تبدیل‌های هندسی به نام تبدیل پروژکتیو استفاده کنیم. تبدیلات پروژکتیو (Projective Transformations) در پردازش تصویر به تغییرات خطی اشاره دارند که شامل تغییر مکان، چرخش، بزرگ‌نمایی، و تغییر پرسپکتیو (زاویه دید) هستند. این تبدیلات باعث می‌شوند تصویر طوری تغییر کند که گویی از زاویه دیگری به آن نگاه می‌کنیم یا دوربین را به شکل خاصی حرکت داده‌ایم.</p>
</div>
</font>

<font face="B Mitra" size=4>
<div dir="rtl">

تبدیلات پروژکتیو با استفاده از یک ماتریس $H$ از فرمول زیر پیروی می‌کنند:

</div>

$$

\begin{bmatrix}
    u' \\
    v' \\
    w'
\end{bmatrix} = H \cdot \begin{bmatrix}
    x \\
    y \\
    1
\end{bmatrix}

 \quad where  \quad

H = \begin{bmatrix}
    h_{11} & h_{12} & h_{13} \\
    h_{21} & h_{22} & h_{23} \\
    h_{31} & h_{32} & h_{33}
\end{bmatrix}
$$

<div dir="rtl">

مختصات نقطه قبل از تبدیل
$(x, y)$
می‌باشد که به فضای مختصات همگن برده شده است و به صورت
$(x,y,1)$
نمایش داده شده است.

لازم است که بردار حاصل از تبدیل یعنی
$('u' ,v' ,w)$
نیز در فضای مختصات همگن نرمال شود (درایه آخر برابر با 1 شود) تا
مختصات نقطه در دستگاه کارتزین به صورت
$(u, v)$
به دست آید. پس به ازای هر نقطه تبدیل شده خواهیم داشت:
$'u'/w = u$
و
$'v'/w = v$


حال فرض کنید که ما 4 نقطه مبدا و 4 نقطه متناظر با آن‌ها در مقصد را داریم و به دنبال ماتریس تبدیلی هستیم که نقاط مبدا را به نقاط مقصد ببرد.
به طور کلی درایه
$h_{33}$
ماتریس تبدیل برابر با 1 در نظر گرفته می‌شود. با این فرض 8 درایه از ماتریس باقی خواهد ماند که باید آن‌ها را پیدا کنیم. در ادامه شما باید با استفاده از روابطی که برای هر نقطه داریم:
</div>


$$

u = \frac{u'}{w'} = \frac{h_{11} x + h_{12} y + h_{13}}{h_{31} x + h_{32} y + 1} \Rightarrow u = h_{11} x + h_{12} y + h_{13} - h_{31} xu - h_{32} yu

$$
$$

v = \frac{v'}{w'} = \frac{h_{21} x + h_{22} y + h_{23}}{h_{31} x + h_{32} y + 1} \Rightarrow v = h_{21} x + h_{22} y + h_{23} - h_{31} xv - h_{32} yv

$$
<div dir="rtl">

دستگاه معادلاتی را تشکیل دهید که با حل آن مقادیر درایه‌های ماتریس
$H$
به دست آید. پس از به دست آوردن ماتریس تبدیل می‌توانیم تبدیل مورد نظرمان را اعمال کنیم.
</div>
</font>

```python
import cv2
import requests
from google.colab.patches import cv2_imshow
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

def read_image_from_url(url, flags=cv2.IMREAD_COLOR):
    response = requests.get(url)
    response.raise_for_status()
    img_array = np.array(bytearray(response.content), dtype=np.uint8)
    img = cv2.imdecode(img_array, flags)
    return img
```

```python
source_points = np.array([[364, 191], [709, 191], [364, 647], [709, 647]], dtype=np.float32)
destination_points = np.array([[179, 890], [532, 827], [591, 966], [930, 877]], dtype=np.float32)
```

<font face="B Mitra" size=4>

<div dir="rtl">
    <h3> بدست آوردن ماتریس تبدیل</h3>
    <p>تابعی بنویسید که یک دستگاه معادلات خطی را تنظیم کند تا ماتریس تبدیل را بر اساس نقاط مبدا و مقصد محاسبه کند. از <code>np.linalg.solve</code> برای حل معادلات استفاده کنید.</p>
</div>
</font>

```python
# Function to calculate the manual transformation matrix from source to destination points
def get_manual_transform(src_pts, dst_pts):
    #TODO Fill here
    A = np.zeros((8, 8))
    b = np.zeros(8)

    for i in range(4):
        x, y = src_pts[i]
        u, v = dst_pts[i]

        # u = h_11.x + h_12.y + h_13 − h_31.x.u − h_32.y.u
        A[2*i] = [x, y, 1, 0, 0, 0, -x * u, -y * u]
        b[2*i] = u

        # v = h_21.x + h_22.y + h_23 − h_31.x.v − h_32.y.v
        A[2*i + 1] = [0, 0, 0, x, y, 1, -x * v, -y * v]
        b[2*i + 1] = v

    # A * H = b
    H = np.linalg.solve(A, b)
    H = np.append(H, 1).reshape(3, 3)

    return H
```

<font face="B Mitra" size=4>

<div dir="rtl">
    <h3> اعمال ماتریس تبدیل</h3>
    <p> در ابتدا ماتریس تبدیل متناظر با نقاط مبدا و مقصد را به دست آورید و سپس تبدیل را بر روی تمام نقاط درون ناحیه مبدا انجام دهید. </p>
</div>
</font>

```python
#apply transformation matrix to source points
  #TODO Fill here

H1 = get_manual_transform(source_points, destination_points)
print(f"Transformation Matrix H1:\n{H1}")

img_height, img_width = image.shape[:2]

# Apply the Perspective Transformation to the Entire Image
warped_image = np.zeros((img_height, img_width, 3), dtype=np.uint8)
for y in range(img_height):
    for x in range(img_width):
        src_point = np.array([x, y, 1])
        dst_point = np.dot(H1, src_point)
        x_new, y_new = int(dst_point[0] / dst_point[2]), int(dst_point[1] / dst_point[2])

        if 0 <= x_new < img_width and 0 <= y_new < img_height:
            warped_image[y_new, x_new] = image[y, x]

# Replace zero values in the transformed image with the mean of non-zero neighboring pixels
corrected_image = warped_image.copy()
for y in range(img_height):
    for x in range(img_width):
        if np.all(warped_image[y, x] == 0):
            non_zero_neighbors = []

            for j in [-1, 0, 1]:
                for i in [-1, 0, 1]:
                    if not (i == 0 and j == 0):
                        y_neighbor, x_neighbor = y + j, x + i
                        if 0 <= y_neighbor < img_height and 0 <= x_neighbor < img_width:
                            if np.any(warped_image[y_neighbor, x_neighbor] != 0):
                                non_zero_neighbors.append(warped_image[y_neighbor, x_neighbor])

            if non_zero_neighbors:
                corrected_image[y, x] = np.mean(non_zero_neighbors, axis=0)

mask = np.zeros((img_height, img_width), dtype=np.uint8)

# Ensure the destination points are ordered correctly (clockwise)
ordered_dst_points = np.array([destination_points[0], destination_points[1], destination_points[3], destination_points[2]], dtype=np.int32)

cv2.fillPoly(mask, [ordered_dst_points], 255) # Fill the Destination Points Area in the Mask with White
mask_bgr = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)

transformed_region = cv2.bitwise_and(corrected_image, mask_bgr)
combined_image = cv2.bitwise_and(image, cv2.bitwise_not(mask_bgr))
combined_image = cv2.bitwise_or(combined_image, transformed_region)
```
